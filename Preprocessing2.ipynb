```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
delhi = pd.read_csv(r"C:\Users\DELL\Downloads\AQI_PROJECT\Delhi_combined_refined.csv", index_col="Timestamp", parse_dates=True)
mumbai = pd.read_csv(r"C:\Users\DELL\Downloads\AQI_PROJECT\Mumbai_combined_refined.csv", index_col="Timestamp", parse_dates=True)
kolkata = pd.read_csv(r"C:\Users\DELL\Downloads\AQI_PROJECT\Kolkata_combined_refined.csv", parse_dates=True)
bengaluru = pd.read_csv(r"C:\Users\DELL\Downloads\AQI_PROJECT\Bengaluru_combined_refined.csv", index_col="Timestamp", parse_dates=True)
```

```python
delhi = delhi.round(2)
```

```python
delhi = delhi.reset_index()
```

```python
delhi = delhi.set_index("Timestamp")
```
